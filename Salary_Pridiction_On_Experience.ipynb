```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error
import gradio as gr
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
salary = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Salary%20Data.csv')

# Split the data into X and y
X = salary[['Experience Years']]
y = salary['Salary']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2529)

# Build the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions and calculate model accuracy (MAPE)
y_pred = model.predict(X_test)
mape = mean_absolute_percentage_error(y_test, y_pred) * 100

# Function to predict salary and display a graph based on experience
def predict_salary(experience):
    experience = float(experience)
    prediction = model.predict([[experience]])[0]

    # Create a graph showing salary vs. experience
    fig, ax = plt.subplots()
    x_vals = np.linspace(0, 20, 100)  # Range of experience years
    y_vals = model.predict(x_vals.reshape(-1, 1))
    ax.plot(x_vals, y_vals, label="Predicted Salary Curve", color="blue")
    ax.scatter(experience, prediction, color='red', label="Your Prediction", s=100)
    ax.set_xlabel('Years of Experience')
    ax.set_ylabel('Salary ($)')
    ax.legend()
    plt.title("Salary vs. Experience")

    # Save and return the graph
    plt.savefig('salary_vs_experience.png')

    return f"Predicted Salary: ${prediction:,.2f}\n\nModel Accuracy (MAPE): {100 - mape:.2f}%", 'salary_vs_experience.png'

# Gradio interface
iface = gr.Interface(
    fn=predict_salary,
    inputs=gr.Number(label="Enter Years of Experience", precision=2),  # Removed placeholder
    outputs=[gr.Textbox(label="Prediction and Model Accuracy", lines=5), gr.Image(label="Salary vs Experience Graph")],
    title="Salary Prediction System",
    description="""
    Welcome to the Salary Prediction System! 🎉\n
    Simply enter your years of experience, and our model will predict your expected salary.\n
    This system uses machine learning to make accurate predictions, and you can also see the model's accuracy along with a visual graph showing the relationship between experience and salary.
    """,
    css=".gradio-container {background-color: #f0f8ff; font-size: 18px; color: #333333; font-family: 'Arial'; padding: 20px;}",
    allow_flagging="never"  # Disables flagging to make the interface cleaner
)

# Launch the Gradio interface
iface.launch(share=True)
```
