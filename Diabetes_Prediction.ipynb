```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
import numpy as np
import gradio as gr

# Load the dataset
diabetes = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/refs/heads/main/Diabetes.csv')

# Prepare features and target variable
y = diabetes['diabetes']
X = diabetes[['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']]

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=2529)

# Train the model
model = LogisticRegression(max_iter=500)
model.fit(X_train, y_train)

# Function to generate random entries for prediction
def generate_random_entries(num_entries=5):
    random_indices = np.random.choice(len(diabetes), num_entries, replace=False)
    entries = diabetes.iloc[random_indices][['pregnancies', 'glucose', 'diastolic', 'triceps', 'insulin', 'bmi', 'dpf', 'age']].values
    results = []

    for entry in entries:
        scaled_data = scaler.transform([entry])
        prediction = model.predict(scaled_data)[0]
        results.append({
            "Pregnancies": entry[0],
            "Glucose": entry[1],
            "Diastolic": entry[2],
            "Triceps": entry[3],
            "Insulin": entry[4],
            "BMI": entry[5],
            "DPF": entry[6],
            "Age": entry[7],
            "Prediction": "Diabetes Positive" if prediction == 1 else "Diabetes Negative"
        })

    return pd.DataFrame(results)

# Gradio Interface
iface = gr.Interface(
    fn=generate_random_entries,
    inputs=None,  # No inputs needed
    outputs=gr.Dataframe(label="Diabetes Prediction Results"),
    title="Diabetes Prediction System",
    description="Random sample entries and their diabetes predictions will be shown each time.",
    theme="default"
)

# Launch the interface
iface.launch(share=True)  # share=True generates a public link
```
